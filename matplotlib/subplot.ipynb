## Subplot

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel(" larges x axes")
axes1.set_ylabel("large y axes")
axes1.set_title("large")

axes2.plot(y,x,'g')
axes2.set_xlabel("x axes")
axes2.set_ylabel("y axes")
axes2.set_title("small")
```

```python
fig,axes=plt.subplots(nrows=1,ncols=10)

for i in axes:
    i.plot(x,y,'g')
    i.set_xlabel("x axes")
    i.set_ylabel("y axes")
    i.set_title("Subplots")
plt.tight_layout(pad=0.2, w_pad=0.3, h_pad=1.0)
```

```python
fig=plt.figure(figsize=(9,4),dpi=100)
```

```python
fig,axes=plt.subplots(figsize=(12,3))
axes.plot(x,y,color="#a7acf1",alpha=0.5)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('ocean')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend(loc=1)
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```
