## Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.linspace(1,10,21)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Example")
plt.show()
```

```python
x=np.array(["A","B","C","D","E"])
y=np.array([2,4,6,8,10])
plt.bar(x,y)
plt.xlabel("Categorical value")
plt.ylabel("Numerical value")
plt.title("Bar Chart")
plt.show()
```

```python
# Histogram plot
x=np.random.normal(30,20,40)
```

```python
x
```

```python
plt.hist(x)
plt.show()
```

```python
# Pie chart
x=np.array([20,40,30,10])
plt.pie(x)
plt.show()
```

```python
# Scatter plot
x=np.array([2,3,5,6,7,8,9,15,12,41,25])
y=np.array([95,85,47,63,21,48,5,36,12,32,12])
plt.scatter(x,y)
plt.show()
```

```python
# Sub plot
x=np.linspace(1,8,15)
y=x**2
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(x,y,"g*-")
plt.show()
```
