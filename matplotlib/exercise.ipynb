```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
x=np.array([1,2,3,4])
y=np.array([3,4,5,6])
axes1.plot(x,y,)
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes2.plot(x,y)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('zoom')
```
