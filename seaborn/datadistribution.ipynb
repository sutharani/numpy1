##  Data Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Data loading
```

```python
# tips = sns.load_dataset("tips")
```

```python
df=sns.load_dataset('tips')
df
```

## Univariate

```python
sns.displot(df["total_bill"],kde=True)
```

```python
# Rug plot
```

```python
sns.rugplot(df["total_bill"])
```

```python
# Kde plot
```

```python
sns.kdeplot(df["total_bill"])
sns.rugplot(df["total_bill"])
```

##  Bivariate

```python
# Join plot
```

```python
sns.jointplot(x=df["total_bill"],y=df["tip"],data=df)
```

```python
sns.jointplot(x=df["total_bill"],y=df["tip"],data=df,kind="reg")
```

## Multivariate

```python
# Pairplot
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex",palette="coolwarm")
```
