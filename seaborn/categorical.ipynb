## Univariate

```python
import seaborn as sns
import numpy as np
```

```python
flights = sns.load_dataset("flights")
df=sns.load_dataset("flights")
df
```

```python
sns.barplot(x='year',y='passengers',data=df)
```

```python
sns.barplot(x='year',y='passengers',data=df,estimator=np.std)
```

```python
# Count plot
```

```python
sns.countplot(x='year',data=df)
```

```python
# Boxplot
```

```python
sns.boxplot(x='year',y='passengers',data=df,palette="rainbow")
```

```python
sns.boxplot(data=df,palette="rainbow")
```

```python
sns.boxplot(data=df,palette="rainbow",orient='h')
```

```python
# Bivariate
```

```python
# Violin plot
```

```python
sns.violinplot(x="year", y="passengers", data=df,palette='rainbow')
sns.swarmplot(x="year", y="passengers", data=df,color='black',size=3)
```
